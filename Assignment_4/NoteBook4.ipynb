```python
import pandas as pd

# Load dataset
df = pd.read_csv('telecom_customer_churn_cleaned.csv')

# Basic statistics
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
mode_values = df.mode(numeric_only=True).iloc[0]
variance_values = df.var(numeric_only=True)

print("Mean:\n", mean_values, "\n")
print("Median:\n", median_values, "\n")
print("Mode:\n", mode_values, "\n")
print("Variance:\n", variance_values, "\n")

# Correlation matrix
correlation = df.corr(numeric_only=True)
print("Correlation Matrix:\n", correlation)
```

Correlation analysis. Identifying the 3 features most related to target variable

```python
# Convert target to numeric (Churned=1, Stayed=0)
df['Customer Status'] = df['Customer Status'].map({'Churned': 1, 'Stayed': 0})

# Correlation with target
corr_with_target = df.corr(numeric_only=True)['Customer Status'].sort_values(ascending=False)
print("Correlation with Target Variable:\n", corr_with_target)

# Top 3 features most correlated
top3 = corr_with_target[1:4]
print("\nTop 3 Features Most Related to Churn:\n", top3)
```
